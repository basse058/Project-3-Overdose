```python
import pandas as pd
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
data = "VSRR_Provisional_Drug_Overdose_Death_Counts.csv"
```

```python
#dropped erroneous columns.
data_df = pd.read_csv(data)
data_df = pd.DataFrame(data_df.rename(columns={"State Name":"state_name", "Data Value":"value", "State":"state", "Year":"year", "Month":"month", "Indicator":"indicator"}))
data_df = data_df[["state", "year", "month", "indicator", "value", "state_name"]]
data_df
```

```python
data_df.replace({'month':{'January':'1', 'February':'2', 'March':'3','April':'4', 'May':'5','June':'6','July':'7','August':'8', 'September':'9', 'October':'10', "November":'11','December':'12'}}, inplace=True)
data_df
```

```python
#dropped non-states from dataset.
list01 = ["New York City","United States"]
data_df = (data_df[data_df.state_name.isin(list01) == False])
data_df
```

```python
#Dropped missing values.
new_df = data_df.dropna()
new_df['value']= new_df['value'].str.replace(',','')
#new_df = pd.DataFrame(new_df.rename(columns={"State_name":"State"}))
new_df
```

```python
#create year/month dataframe
time_df = new_df.drop_duplicates(["year", "month"]).reset_index(drop=True)[["year", "month"]]
time_df = pd.DataFrame(time_df)
time_df
```

```python
#create state dataframe
state_df = new_df.drop_duplicates(["state"]).reset_index(drop=True)[["state"]]
state_df = pd.DataFrame(state_df)
```

```python
#
od_death_data = new_df.loc[new_df['indicator'] == 'Number of Drug Overdose Deaths']
od_death_data.rename(columns={'value': 'overdose_deaths'}, inplace=True)
od_death_data['year'] = od_death_data['year'].astype(str)
od_death_data.dtypes
```

```python
od_death_data["Date_Month_Year"] = od_death_data["month"] + " " + od_death_data["year"]
# Now we will parse the Date so that the notebook can read it as an actual date
import datetime
od_death_data['date'] = pd.to_datetime(od_death_data['Date_Month_Year'], format="%B %Y")
od_death_data
```

```python
date_indexed_opioid_data = od_death_data.set_index('date')
# Taking a look
date_indexed_opioid_data.head()
```

```python
od_death_data.to_csv('od_data_df.csv', index=True)
```

```python
#
death_data = new_df.loc[new_df['Indicator'] == 'Number of Deaths']
death_data.rename(columns={'Value': 'Total_deaths'}, inplace=True)
death_data['Year'] = death_data['Year'].astype(str)
death_data
```

```python
death_data["Date_Month_Year"] = death_data["Month"] + " " + death_data["Year"]
# Now we will parse the Date so that the notebook can read it as an actual date
import datetime
death_data['Date'] = pd.to_datetime(death_data['Date_Month_Year'], format="%B %Y")
death_data
```

```python
date_indexed_death_data = death_data.set_index('Date')
# Taking a look
date_indexed_death_data.head()
```

```python
death_data.to_csv('../Resources/death_data_df.csv', index=True)
date_indexed_death_data.to_csv('../Resources/death_index.csv', index=False)
```

```python
#
percent_data = new_df.loc[new_df['Indicator'] == 'Percent with drugs specified']
percent_data
```

```python
time_df.to_csv('Resources/time_df.csv', index=False)
state_df.to_csv('Resources/state_df.csv', index=False)
od_death_data.to_csv('Resources/od_death_data_df.csv', index=False)
death_data.to_csv('Resources/death_data_df.csv', index=False)
percent_data.to_csv('Resources/percent_df.csv', index=False)
```
